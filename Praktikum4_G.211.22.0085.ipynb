```python
def create_pangram():
    pangram = ""  # Langkah 1: Buat string kosong untuk pangram
    used_letters = set()  # Langkah 2: Buat set untuk melacak huruf yang sudah digunakan

    alphabet = "abcdefghijklmnopqrstuvwxyz"  # Alfabet lengkap

    for letter in alphabet:  # Langkah 3: Iterasi melalui huruf-huruf dalam alfabet
        if letter not in used_letters:  # Langkah 4: Periksa apakah huruf belum digunakan
            pangram += letter  # Tambahkan huruf ke pangram
            used_letters.add(letter)  # Tandai huruf sebagai digunakan

    return pangram  # Langkah 6: Kembalikan pangram yang telah dibuat

# Contoh penggunaan
pangram = create_pangram()
print("Pangram yang dibuat:", pangram)
```

```python
def is_pangram(s):
    # Langkah 1: Buat vektor tanda (mark) boolean dengan ukuran 26
    mark = [False] * 26

    # Konversi semua huruf dalam string ke huruf kecil
    s = s.lower()

    # Langkah 2: Tandai huruf yang ada dalam string
    for char in s:
        if char.isalpha():
            # Tandai huruf yang valid
            mark[ord(char) - ord('a')] = True

    # Langkah 3: Periksa apakah semua indeks ditandai (semua karakter ada)
    return all(mark)

# Contoh penggunaan
input_string = "The quick brown fox jumps over the lazy dog"
result = is_pangram(input_string)

if result:
    print("Ini adalah pangram.")
else:
    print("Ini bukan pangram.")
```
